```python
from collections import Counter
(pd.Series(item_categories.item_category_name.str.replace(r"-|,|\)|\(","").str.split()
           .apply(Counter).sum()).sort_values(ascending=False)
)
```

```python
item_categories[~item_categories.item_category_name.str.contains("-")]
```

```python
(item_categories
 [item_categories.item_category_name.str.contains("-")]
 .item_category_name
 .str.split(" - ")
 .to_frame()
 .assign(category=lambda x: x.item_category_name.apply(lambda y: y[0]))
.assign(subcategory=lambda x: x.item_category_name.apply(lambda y: y[1]))
)
```
